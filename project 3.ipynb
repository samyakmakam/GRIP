```python
plt.figure(figsize=(10,5))
sns.countplot(d['State'])
plt.xticks(rotation=90);
```

```python
dff=pd.Series(d['State'])
dff=pd.DataFrame(dff)
dff['Category']=d['Category']
dff['Quantity']=d['Quantity']
dff.set_index('Category')
dff.reindex()
```

```python
plt.figure(figsize=(50,25))
df_fur=dff.loc[dff.Category=='Furniture']
df0=pd.DataFrame({'x':df_fur['State'],'y':df_fur['Quantity']})
df_tec=dff.loc[dff.Category=='Technology']
df6=pd.DataFrame({'x':df_tec['State'],'y':df_tec['Quantity']})
df_o=dff.loc[dff.Category=='Office Supplies']
df7=pd.DataFrame({'x':df_o['State'],'y':df_o['Quantity']})
df0['hue']='Furniture'
df6['hue']='Technology'
df7['hue']='Offline Supplies'
res=pd.concat([df0,df6,df7])
sns.barplot(x='x',y='y',data=res,hue='hue')
plt.xticks(rotation=90,fontsize=35)
plt.yticks(fontsize=35)
plt.legend(fontsize=35)
plt.show()
```

```python
df12=d.groupby('City')['Profit'].sum().reset_index().sort_values(by='Profit',ascending=False)
df13=df12.head(10)
plt.figure(figsize=(50,25))
plt.bar(df13['City'],df13['Profit'])
plt.xticks(rotation=90,fontsize=35)
plt.yticks(fontsize=35)
plt.show()
```

```python
df17=d.groupby('City')['Sales'].sum().reset_index().sort_values(by='Sales',ascending=False)
df18=df17.head(10)
plt.figure(figsize=(50,25))
plt.bar(df18['City'],df18['Sales'])
plt.xticks(rotation=90,fontsize=35)
plt.yticks(fontsize=35)
plt.show()
```

```python
df12=d.groupby('City')['Profit'].sum().reset_index().sort_values(by='Profit',ascending=False)
df13=df12.tail(10)
plt.figure(figsize=(50,25))
plt.bar(df13['City'],df13['Profit'])
plt.xticks(rotation=90,fontsize=35)
plt.yticks(fontsize=35)
plt.show()
```

```python
df17=d.groupby('City')['Sales'].sum().reset_index().sort_values(by='Sales',ascending=False)
df18=df17.tail(10)
plt.figure(figsize=(50,25))
plt.bar(df18['City'],df18['Sales'])
plt.xticks(rotation=90,fontsize=35)
plt.yticks(fontsize=35)
plt.show()
```
