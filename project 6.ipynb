# Decision Tree

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
iris=load_iris()
X=iris.data
y=iris.target
tree=DecisionTreeClassifier(max_depth=4,random_state=0)
tree.fit(X,y)
```

```python
y
```

```python
iris.target_names
```

```python
from sklearn.tree import export_graphviz
export_graphviz(tree, out_file='three.dot',class_names=['setosa', 'versicolor', 'virginica'], feature_names=iris.feature_names,filled=True, rounded=True,special_characters=True)
```
